```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
# Loading data set
iris = datasets.load_iris()
```

```python
print(iris.keys())
```

```python
#Type of the data
type(iris), type(iris.data), type(iris.target)
```

```python
#Shape of the data (contain the variables)
iris.data.shape
```

```python
# The name of the classes in the target (index: start with 0)
iris.target_names
```

#### target names and classes are
- 'setosa'=0,
- 'versicolor'=1, 
- 'virginica'=2

```python
iris.feature_names
```

```python
X = iris.data
y = iris.target
df = pd.DataFrame(X, columns = iris.feature_names)
print(df.head(2))
```

```python
# visualize
_ = pd.plotting.scatter_matrix(df, c = y, figsize = [8,8],\
                              s = 10, marker = 'D')
```

```python
# SciKit-Learn fit and predict method
# training a model on data using .fit() method and predicting the labels of new data by .predict() method
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 6)
```

```python
# Fitting the model on train data
knn.fit(iris['data'], iris['target'])
```

#### REQUIRMENTS of SciKit Learn API:
- data as numpy array or packaged data frame.
- No missing values.
- Continuous variable in feature
- Every column is a feature in data

```python
#iris.data
```

### Testing model on new data

```python
# Testing with unkown data

X_new = np.array([[5.1, 3.5, 1.4, 0.2],
       [4.9, 3. , 3.4, 0.2],
       [6.7, 3. , 5.2, 2.3]])
```

```python
prediction = knn.predict(X_new)
print('Prediction: {}'.format(prediction))
print('0:Setosa ; 1:Versicolor; 2:Virginica')
```
